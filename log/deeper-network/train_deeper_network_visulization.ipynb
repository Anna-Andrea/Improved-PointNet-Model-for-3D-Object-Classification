```python
# 创建空列表存储每个epoch的eval mean loss，eval accuracy和eval avg class acc
eval_mean_loss_list = []
eval_accuracy_list = []
eval_avg_class_acc_list = []
epochs = []

# 打开文件并逐行读取内容
with open('log_train.txt', 'r') as file:
    lines = file.readlines()
    
    for line in lines:
        # 拆分每行内容
        words = line.split()
        '''
        ['eval', 'mean', 'loss:', '1.605335']
        ['eval', 'accuracy:', '0.547484']
        ['eval', 'avg', 'class', 'acc:', '0.457884']
        '''
        if words[0] == 'eval' and words[1] == 'mean' and words[2] == 'loss:':
            eval_mean_loss_list.append(float(words[-1]))
        if words[0] == 'eval' and words[1] == 'accuracy:':
            eval_accuracy_list.append(float(words[-1]))
        if words[0] == 'eval' and words[1] == 'avg' and words[2] == 'class' and words[3] == 'acc:' :
            eval_avg_class_acc_list.append(float(words[-1]))

# 打印列表
epochs = list(range(len(eval_mean_loss_list)))
print("epochs:",epochs)
print("Eval Mean Loss:", eval_mean_loss_list)
print("Eval Accuracy:", eval_accuracy_list)
print("Eval Avg Class Acc:", eval_avg_class_acc_list)
```

```python
import matplotlib.pyplot as plt

# 将文本内容转换为可处理的数据

# 绘制曲线图
plt.figure(figsize=(10, 6))
plt.plot(epochs, eval_mean_loss_list, marker='o', label='Eval Mean Loss')
plt.plot(epochs, eval_accuracy_list, marker='o', label='Eval Accuracy')
plt.plot(epochs, eval_avg_class_acc_list, marker='o', label='Eval Avg Class Acc')

# 添加标题和标签
plt.title('Evaluation Metrics vs. Epochs (PointNet_SE_BN_Deeper)')
plt.xlabel('Epochs')
plt.ylabel('Metrics Value')
plt.xticks(epochs)
plt.grid(True)
plt.legend()
plt.xticks(range(0, 240, 10))  # 每隔20个epoch显示一个刻度

# 显示图表
plt.tight_layout()
plt.show()
```
